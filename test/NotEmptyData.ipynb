```python
slope = 2.666051223553066e-05
hover_text = []
bubble_size = []
```

```python
for index,row in df_2007.iterrows():
    hover_text.append(('Location: {location}<br>'+ 'Year :{year}<br>'+ 'Primary Type :{primary_type}').format(location=row['Location'],year=row['Year'],primary_type=row['Primary Type']))
    bubble_size.append(math.sqrt(row['ID']*slope))
df_2007['text'] = hover_text
df_2007['size'] = bubble_size
sizeref = 2.*max(df_2007['size'])/(100**2)
```
