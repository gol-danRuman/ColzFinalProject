```python
import dateutil.parser
divisor = 1000000000.0
epochDate = []
for d in samples['Date']:
    date = dateutil.parser.parse(d, dayfirst=True).timestamp() / divisor 
    epochDate.append(date)
epochDate[9]
```

```python
arrest = samples['Arrest']
```

```python
falseFreq = arrest.describe().freq
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Arrest', 'NotArrested'
sizes = [arrest.count()-falseFreq, falseFreq]
explode = (0,  0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
fig1.savefig('ArrestDataPieChart.png',dpi=100)
```
