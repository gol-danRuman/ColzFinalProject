```python
#convert date object to list of month,year,hour etc
from datetime import datetime

date_list = df['Date'].tolist()
block_list = df['Block'].tolist()
primarytype_list = df['Primary Type'].tolist()
description_list = df['Description'].tolist()
convertdate_List = [] #empty list
month_list = []
days_list = []
year_list = []
hour_list = []

for date_str1 in date_list: 
    date = datetime.strptime(date_str1, '%m/%d/%Y %H:%M:%S %p')
    convertdate_List.append(date)
    month_list.append(date.month)
    days_list.append(date.day)
    year_list.append(date.year)
    hour_list.append(date.hour)
```

```python
#Append list to existing dataframe
Month = pd.DataFrame(month_list)
df = df.assign(Month=Month.values)
```
